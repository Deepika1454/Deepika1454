```python
!pip install transformers torch

from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import zipfile
import os

# Specify the path to your zip file
zip_file_path = '/content/archive (1).zip'

# Specify the directory where you want to extract the model
extract_dir = '/content/extracted_model'

# Check if the file exists
if not os.path.exists(zip_file_path):
    print(f"Error: Zip file not found at {zip_file_path}. Please verify the path.")
else:
    # Check if the file is a valid zip file
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            # Extract the model from the zip file
            zip_ref.extractall(extract_dir)
            print(f"Model extracted to {extract_dir}")

            # Load the tokenizer and model from the extracted directory
            tokenizer = AutoTokenizer.from_pretrained(extract_dir)
            model = AutoModelForSequenceClassification.from_pretrained(extract_dir)

            # Create the sentiment analysis pipeline
            sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
    except zipfile.BadZipFile:
        print(f"Error: Invalid zip file at {zip_file_path}. Please re-download or check for corruption.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
from transformers import pipeline

# Load the pre-trained sentiment analysis model
sentiment_analyzer = pipeline('sentiment-analysis')

# Sample texts for analysis
texts = [
    "I love this product! It's amazing!",
    "I hate the service, it was terrible.",
    "I HATE YOU BUT I LOVE YOU"
    "TGIF is right...I think I broke my toe last night - on my bad foot."
]

# Analyze sentiment of each text
for text in texts:
    result = sentiment_analyzer(text)
    print(f"Text: {text}\nSentiment: {result[0]['label']}\n")
```
